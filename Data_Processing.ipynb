```python
import pandas as pd
import numpy as np
import seaborn as snb
import matplotlib as plt
```

```python
car_data = pd.read_csv("car.csv")
car_data_2 = pd.read_csv("car_data.csv")
```

```python
car_data["Hızlanma (0-100/Sn)"] = car_data_2["Hızlanma (0-100/Sn)"]
car_data["Agirlik"] = car_data_2["Agirlik"]
car_data["Silindir_Sayisi"] = car_data_2["Silindir_Sayisi"]
```

```python
car_data.drop(["İlan No"], inplace=True,axis=1)
```

```python
car_data.head()
```

```python
#Fixing Fiyat Column
```

```python
car_data["Fiyat"] = car_data["Fiyat"].str.replace(" TL","")
```

```python
car_data["Fiyat"] = car_data["Fiyat"].str.replace(".","")
```

```python
car_data["Fiyat"] = car_data["Fiyat"].str.replace('\r\n', '').str.replace('\n', '').str.replace(' ', '').astype(float)
```

```python
car_data
```

```python
pd.DataFrame(pd.DataFrame(car_data["Marka"].value_counts()).index)
```

```python
# Fixing Yıl Column
```

```python
car_data["Yıl"].value_counts
```

```python
car_data = car_data[car_data["Yıl"] != "23.000 km"]
```

```python
car_data["Yıl"] = car_data["Yıl"].astype(int)
```

```python
car_data.head()
```

```python
#Fixing Kilometre Column
```

```python
car_data["Kilometre"] = car_data["Kilometre"].str.replace(" km","").str.replace(".","").astype(float)
```

```python
# Fixing Kasa Tipi Column
```

```python
car_data["Kasa Tipi"].value_counts()
```

```python
allowed_values = ["Sedan", "Hatchback/5", "Station wagon", "Hatchback/3", "MPV"]
car_data.loc[~car_data["Kasa Tipi"].isin(allowed_values), "Kasa Tipi"] = "Bilinmiyor"
```

```python
car_data["Kasa Tipi"].value_counts()
```

```python
# Fixing Renk Column
```

```python
car_data["Renk"].value_counts()
```

```python
pd.DataFrame(pd.DataFrame(car_data["Renk"].value_counts()).index)
```

```python
terms_to_replace = ["Coupe", "Sedan", "Hatchback/5", "Cabrio", "cc"]
car_data['Renk'] = car_data['Renk'].apply(lambda x: 'Diğer' if any(term in x for term in terms_to_replace) else x)
```

```python
pd.DataFrame(pd.DataFrame(car_data["Renk"].value_counts()).index)
```

```python
# Fixing Motor Hacmi Column
```

```python
car_data["Motor Hacmi"].value_counts()
```

```python
pd.DataFrame(pd.DataFrame(car_data["Motor Hacmi"].value_counts()).index).tail(30)
```

```python
pd.DataFrame(pd.DataFrame(car_data["Motor Hacmi"].value_counts()).index).head(30)
```

```python
kelime = 'cc'
car_data['Motor Hacmi'] = car_data['Motor Hacmi'].apply(lambda x: 'Bilinmiyor' if kelime not in x else x)
```

```python
pd.DataFrame(pd.DataFrame(car_data["Motor Hacmi"].value_counts()).index).head(30)
```

```python
car_data.head()
```

```python
car_data = car_data[car_data["Motor Hacmi"] != "Bilinmiyor"]
```

```python
car_data["Motor Hacmi"] = car_data["Motor Hacmi"].str.replace(" cc","").astype(int)
```

```python
# Fixing Motor Gücü Column
```

```python
car_data["Motor Gücü"].value_counts()
```

```python
car_data = car_data[car_data["Motor Gücü"].str.contains(" hp", case=False, na=False)]
```

```python
car_data = car_data[~car_data["Motor Gücü"].str.contains("-", na=False)]
```

```python
car_data["Motor Gücü"] = car_data["Motor Gücü"].str.replace(" hp","").astype(int)
```
